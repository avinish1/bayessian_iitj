```python
data_config = load_config("network_config.yaml")
```

```python
data = {}
```

```python
import numpy as np
num_samples=1000
# Step 1: Generate independent (non-conditional) variables
for var, props in data_config["variables"].items():
    if "conditional_on" not in props:
        if props["distribution"] == "normal":
            mean = props["parameters"]["mean"]
            std_dev = props["parameters"]["std_dev"]
            data[var] = np.random.normal(loc=mean, scale=std_dev, size=num_samples)
        elif props["distribution"] == "uniform":
            low = props["parameters"]["low"]
            high = props["parameters"]["high"]
            data[var] = np.random.randint(low, high + 1, size=num_samples)

# Step 2: Generate conditional variables
for var, props in data_config["variables"].items():
    if "conditional_on" in props:
        # Get the parent variables' values from the data
        parent_vars = [i.strip() for i in props["conditional_on"].split(",")]
        parent_values = {parent: data[parent] for parent in parent_vars}
        
        # Evaluate the equation for the current variable
        equation = props["equation"]
        
        # Substitute parent values in the equation
        for parent, values in parent_values.items():
            equation = equation.replace(parent, f"parent_values['{parent}']")
        
        # Calculate the current variable using the equation and random noise (std_dev)
        calculated_values = eval(equation)
        
        if "std_dev" in props:
            noise = np.random.normal(0, props["std_dev"], num_samples)
            data[var] = calculated_values + noise
        else:
            data[var] = calculated_values
```

```python
# Step 3: Convert the data to a DataFrame and save it to CSV
import pandas as pd
df = pd.DataFrame(data)
```

```python
df.sort_values(by='research_output', ascending=False)
```

```python
from pgmpy.estimators import MmhcEstimator

# Load the generated CSV data
#data = pd.read_csv("sample_data.csv")

# Use MmhcEstimator to learn the structure
estimator = MmhcEstimator(df)
model = estimator.estimate()
```

```python
model_daft = model.to_daft(
                                  pgm_params={'grid_unit': 10},
                                  node_params={'research_output': {'shape': 'rectangle'}})
# To open the plot
model_daft.render()
```

```python
factors = {}
for var, properties in data_config['variables'].items():
    if properties['distribution'] == 'normal':
        # Independent variable with normal distribution
        mean = properties['parameters']['mean']
        std_dev = properties['parameters']['std_dev']
        factors[var] = lambda mean=mean, std_dev=std_dev: np.random.normal(mean, std_dev)
    elif properties['distribution'] == 'conditional':
        # Conditional variable with equation
        equation = properties['equation']
        std_dev = properties.get('std_dev', 1)
        conditional_vars = properties['conditional_on'].split(", ")
        
        # Create a sampling function using the equation and conditional variables
        def make_conditional_pdf(eq, cond_vars, std_dev):
            return lambda **inputs: np.random.normal(eval(eq, {}, inputs), std_dev)
        
        # Build the sampling function with evaluated equation
        factors[var] = make_conditional_pdf(equation, conditional_vars, std_dev)
```

```python
def generate_samples(factors, num_samples=1000):
    samples = pd.DataFrame()
    
    # Sample independent variables first
    for var, func in factors.items():
        if 'conditional_on' not in data_config['variables'][var]:
            samples[var] = [func() for _ in range(num_samples)]
    
    # Sample conditional variables based on their parent variables
    for var, func in factors.items():
        if 'conditional_on' in data_config['variables'][var]:
            conditional_vars = data_config['variables'][var]['conditional_on'].split(", ")
            samples[var] = [
                func(**{parent: samples.loc[i, parent] for parent in conditional_vars})
                for i in range(num_samples)
            ]
    
    return samples
```

```python
sample_data = generate_samples(factors)
```
